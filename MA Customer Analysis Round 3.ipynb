```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
import seaborn as sns
import scipy.stats as stats


df = pd.read_csv("https://raw.githubusercontent.com/Belphegorus86/lab-customer-analysis-round-3/master/files_for_lab/csv_files/marketing_customer_analysis.csv")
df                 
```

```python
df.info() 
```

```python
#seams like we have no Nan's but lets make sure
df.isna().sum() #confirmed, no nan's
```

```python
df.describe()
```

```python
sns.countplot(data=df, x="Response")
plt.title('Total Number of Responses')
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.countplot(x='Sales Channel', hue='Response', data=df)
plt.title('Response Rate by Sales Channel')
plt.xlabel('Sales Channel')
plt.ylabel('Count')
plt.show()
```

```python
plt.scatter(df['Response'], df['Total Claim Amount'])
plt.title('Rate by Total Claim Amount')
plt.show()
```

```python
plt.scatter(df['Response'], df['Income'])
plt.title('Rate by Income')
plt.show()
```
